```python
# ƒê·ªçc v√†o DataFrame
df = pd.read_csv(url)
df.columns = df.columns.str.strip().str.replace(" ", "_")

display(df.head())
```

### Preprocessing data Structure

```python
df = df.dropna(how='all', axis=0)  # Drop rows where all elements are NaN
df = df.dropna(how='all', axis=1)  # Drop columns where all elements are NaN
df = df.drop(columns='STT', axis=1)  # Drop 'STT' column if it exists
df.info()
```

```python
# Translate column names from Vietnamese to English
mapping = {
    "T√™n_h·ªçc_b·ªïng": "Scholarship_Name",
    "Lo·∫°i_h·ªçc_b·ªïng": "Scholarship_Type",
    "Qu·ªëc_gia": "Country",
    "M·ª©c_t√†i_tr·ª£": "Funding_Level",
    "Chi_ti·∫øt_t√†i_tr·ª£": "Funding_Details",
    "H√¨nh_th·ª©c_m·ªü": "Application_Mode",
    "Th√°ng_m·ªü_ƒë∆°n_(v·ªõi_th∆∞·ªùng_ni√™n)": "Application_Month",
    "Ng√†y_b·∫Øt_ƒë·∫ßu": "Start_Date",
    "Ng√†y_k·∫øt_th√∫c": "End_Date",
    "S·ªë_l∆∞·ª£ng_(n·∫øu_c√≥)": "Quantity",
    "C√≥_d√†nh_cho_sinh_vi√™n_Vi·ªát_Nam_kh√¥ng": "For_Vietnamese",
    "Gi·ªõi_h·∫°n_ƒë·ªëi_t∆∞·ª£ng": "Eligibility",
    "Gi·ªõi_h·∫°n_ng√†nh_ngh·ªÅ": "Field_Restriction",
    "Danh_s√°ch_ng√†nh_ngh·ªÅ_(n·∫øu_c√≥)": "Eligible_Fields",
    "B·∫±ng_c·∫•p_y√™u_c·∫ßu": "Required_Degree",
    "GPA_t·ªëi_thi·ªÉu": "Min_GPA",
    "Y√™u_c·∫ßu_v·ªÅ_X·∫øp_h·∫°ng_c·ªßa_tr∆∞·ªùng_C·ª≠_nh√¢n": "Bachelor_Ranking_Requirement",
    "Y√™u_c·∫ßu_s·ª±_li√™n_quan_c·ªßa_ng√†nh_h·ªçc_c·ª≠_nh√¢n": "Bachelor_Field_Relevance",
    "Y√™u_c·∫ßu_s·ªë_nƒÉm_kinh_nghi·ªám": "Experience_Years",
    "Y√™u_c·∫ßu_s·ªë_gi·ªù_l√†m_vi·ªác_t·ªëi_thi·ªÉu": "Min_Working_Hours",
    "Y√™u_c·∫ßu_ch·ª©ng_ch·ªâ_ngo·∫°i_ng·ªØ": "Language_Certificate",
    "Y√™u_c·∫ßu_ch·ª©ng_ch·ªâ_Academic": "Academic_Certificate",
    "Y√™u_c·∫ßu_gi·∫£i_th∆∞·ªüng,_th√†nh_t√≠ch_h·ªçc_thu·∫≠t": "Awards_Requirement",
    "Y√™u_c·∫ßu_publication": "Publication_Requirement",
    "ƒê·ªô_tu·ªïi": "Age",
    "Gi·ªõi_t√≠nh": "Gender",
    "Ho√†n_c·∫£nh_ƒë·∫∑c_bi·ªát": "Special_Circumstances",
    "K·∫ø_ho·∫°ch_ngh·ªÅ_nghi·ªáp_sau_t·ªët_nghi·ªáp": "Career_Plan",
    "Y√™u_c·∫ßu_Kh√°c": "Other_Requirements",
    "Th√¥ng_tin_h·ªçc_b·ªïng": "Scholarship_Info",
    "H·ªì_s∆°": "Application_Documents",
    "Timeline": "Timeline",
    "URL": "URL",
    "Checked": "Checked"
}


# normalize column names to Capitalized_Words format
def normalize_col(name):
    parts = name.split("_")
    return "_".join([p.capitalize() for p in parts])

# Convert DataFrame
df = df.rename(columns=mapping)
df.columns = [normalize_col(c) for c in df.columns]

display(df.tail())
```

Trong DataFrame ƒëang c√≥ c√°c c·ªôt d·ªØ li·ªáu d·∫°ng chu·ªói (text), nh∆∞ng th·ª±c t·∫ø th√¨ n√≥ c√≥ **2 lo·∫°i t√≠nh ch·∫•t kh√°c nhau**:

   * Sau n√†y n·∫øu c·∫ßn x·ª≠ l√Ω NLP th√¨ t√°ch token, vectorize (TF-IDF, Word2Vec, BERT embedding‚Ä¶), ho·∫∑c t√≥m t·∫Øt n·ªôi dung.

```python
category_cols = ['Scholarship_Type', 
                 'Country', 
                 'Funding_Level', 
                 'Application_Mode', 
                 'Application_Month', 
                 'Quantity',
                 'For_Vietnamese', 
                 'Field_Restriction', 
                 'Required_Degree',
                 'Min_Gpa', 
                 'Bachelor_Field_Relevance',
                 'Language_Certificate',
                 'Academic_Certificate',
                 "Awards_Requirement",
                 "Publication_Requirement",
                 "Age",
                 "Gender",
                 "Other_Requirements",
                 "Checked"
]

for col in category_cols:
    df[col] = df[col].astype('category')
```

```python
real_number_cols = [
    'Experience_Years', 
    'Min_Working_Hours',
]

for col in real_number_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
date_time_cols = [
    'Start_Date', 
    'End_Date',
]

for col in date_time_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
string_cols = [col for col in df.columns if col not in category_cols + real_number_cols + date_time_cols]
for col in string_cols:
    df[col] = df[col].astype('string')
```
