# Edinburgh Global Undergraduate Mathematics Scholarships: Comprehensive Analytical Report

## Basic Information

The Edinburgh Global Undergraduate Mathematics Scholarships are prestigious awards offered by the University of Edinburgh, United Kingdom. These scholarships support international students pursuing undergraduate studies in Mathematics.

- **Official Name:** Edinburgh Global Undergraduate Mathematics Scholarships  
- **Provider:** University of Edinburgh  
- **Study Country:** United Kingdom  
- **Target Audience:** International applicants with International/EU fee status  

## Funding and Benefits

The scholarship provides significant financial assistance, primarily aimed at covering living expenses.

- **Funding Amount:** £5,000 per year  
- **Duration:** Entire duration of the undergraduate degree (subject to satisfactory academic progress)  
- **Covered Expenses:** Paid directly to students in instalments as a contribution towards living costs  

## Timeline and Key Dates

- **Application Deadline:** March 31, 2025 (for the 2025/2026 academic session)  
- **Opening Date:** Applications for the 2025 cycle are open  
- **Number of Awards:** 2 scholarships annually  

## Eligibility Criteria

Applicants must meet the following criteria:

- **Fee Status:** International/EU  
- **Academic Performance:** Outstanding academic records; competitive merit-based selection  
- **Programme of Study:** Full-time undergraduate programme in the School of Mathematics, University of Edinburgh  
- **Exclusions:**  
  - Students already enrolled at the University  
  - Students studying Mathematics as part of a degree hosted by another School  
- **US Student Requirement:** Minimum GPA of 3.0  
- **Required Degree Level:** Undergraduate/Bachelor’s  

## Field of Study Restrictions

- **Eligible Fields:** Undergraduate Mathematics programmes offered by the School of Mathematics  
- **Restrictions:** Mathematics programmes hosted by other Schools are not eligible  

## Application Process and Requirements

- **University Admission:** Apply for full-time undergraduate study via UCAS  
- **Scholarship Application Access:** Must apply for admission before accessing the scholarship portal (offer not required to start)  
- **Selection Process:** Competitive and merit-based  
- **Other Requirements:** No explicit requirements for references, career plans, age, gender, or special circumstances
